```python
import pandas as pd
import numpy as np

from datetime import datetime
```

```python
df = pd.read_csv("iot_telemetry_data.csv")
df.head()
```

```python
group = df.groupby("device")
for device, data in group:
    print(device,data , sep="\n")
```

```python
device_1 = group.get_group("00:0f:00:70:91:0a")
device_1
```

```python
device_1["Datetime"] = pd.to_datetime(device_1['ts'],unit='s')
device_1
```

```python
device_1.set_index("Datetime", inplace=True)
```

```python
s = device_1.resample("Min").mean()
s
```

```python
s.to_csv("data_device1.csv")
```
